## The Concept of Lift

$$Lift(A\rightarrow B):= \frac{\text{P}(B|A)}{\text{P}(B)}$$

The confidence of this rule is the numerator of the quotient representing the Lift:

$$Confidence(A\rightarrow B):= \text{P}(B|A) = \frac{\text{P}(A\cap B)}{\text{P}(A)}$$

We obtain,


$$ Lift(A\rightarrow B)= \frac{\text{P}(A\cap B)}{\text{P}(A)\cdot \text{P}(B)}$$
