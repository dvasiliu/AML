#### Introduction to Classes in Python

In Python, a class is a blueprint for creating objects. Objects are instances of classes, and classes define the properties (attributes) and behaviors (methods) that the objects created from the class can have. Classes are a fundamental concept in object-oriented programming (OOP) and allow for the creation of complex data structures and models.

#### Key Concepts

1. **Class**: A blueprint for creating objects. It defines a set of attributes and methods that the created objects can have.

2. **Object**: An instance of a class. Each object can have unique values for the attributes defined in the class.

3. **Attributes**: Variables that belong to a class or an instance of a class. They represent the properties or state of an object.

4. **Methods**: Functions defined within a class that describe the behaviors of the objects created from the class.

5. **Inheritance**: A way to form new classes using classes that have already been defined. It allows for the creation of a class that is a modified version of an existing class.

6. **Encapsulation**: The practice of keeping the details of the implementation hidden, only exposing the necessary parts of the code.

7. **Polymorphism**: The ability to present the same interface for different underlying data types.

#### Creating a Class

To create a class in Python, use the `class` keyword followed by the class name and a colon. Inside the class, you can define attributes and methods.

```python
class MyClass:
    # Class attribute
    class_attribute = "I am a class attribute"

    # Initializer / Instance attributes
    def __init__(self, instance_attribute):
        self.instance_attribute = instance_attribute

    # Method
    def display_attributes(self):
        print(f"Class Attribute: {MyClass.class_attribute}")
        print(f"Instance Attribute: {self.instance_attribute}")

# Creating an instance of MyClass
my_object = MyClass("I am an instance attribute")

# Accessing attributes and methods
my_object.display_attributes()
```

#### Explanation

- **Class Definition**: The `class MyClass:` line defines a new class named `MyClass`.
- **Class Attribute**: `class_attribute` is a class attribute shared by all instances of the class.
- **Initializer Method**: The `__init__` method is a special method called a constructor. It is called when an object is created from the class and allows the class to initialize the attributes of the object.
- **Instance Attribute**: `self.instance_attribute` is an instance attribute, unique to each instance of the class.
- **Method**: The `display_attributes` method is a function defined within the class that can be called on instances of the class.
- **Creating an Instance**: `my_object = MyClass("I am an instance attribute")` creates an instance of `MyClass` with a unique instance attribute.
- **Accessing Methods and Attributes**: `my_object.display_attributes()` calls the `display_attributes` method, which prints the class and instance attributes.

#### **Dunder Methods**

Dunder methods, also known as "magic methods" or "special methods," are methods in Python classes that have names surrounded by double underscores (dunder for "double underscore").

**Definition:**

Dunder methods are special methods in Python that allow you to define or customize how your objects interact with built-in Python operators, functions, and language constructs. They are identified by their distinctive naming convention of having two leading and two trailing underscores, like `__init__`, `__str__`, `__add__`, etc.

In simpler terms:

They are methods that Python automatically calls behind the scenes in response to specific operations or events. They let you hook into Python's internal machinery and define how your custom objects should behave in various situations.

**Characteristics of Dunder Methods:**

Naming Convention: Always start and end with two underscores (__).
Implicit Calls: You rarely call them directly. Python calls them implicitly based on context. For example:

- `__init__` is called when you create a new object instance.

- `__str__` is called when you use print(your_object) or str(your_object).

- `__add__` is called when you use the + operator with your object (e.g., obj1 + obj2).

**Customization:** They allow you to override default behaviors and tailor how your objects work.

**Operator Overloading:** They enable you to define how operators (+, -, *, /, ==, !=, <, >, etc.) work with your objects.
Integration with Built-in Functions: They make your objects compatible with built-in functions like len(), str(), repr(), abs(), etc.
Purpose:

The primary purpose of dunder methods is to give you a great deal of control over the behavior of your objects, making them seamlessly integrate with the Python language and its features. They help you create more expressive, intuitive, and powerful classes.

Example:

```python
class MyNumber:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        if isinstance(other, MyNumber):
            return MyNumber(self.value + other.value)
        else:
            return MyNumber(self.value + other)

    def __str__(self):
        return f"MyNumber({self.value})"


num1 = MyNumber(5)
num2 = MyNumber(10)
num3 = num1 + num2  # __add__ is called here
print(num3)        # __str__ is called here (Output: MyNumber(15))
```

In this example:

`__init__` is the dunder method that initializes the object.

`__add__` is the dunder method that defines what happens when you use the + operator with MyNumber objects.

`__str__` defines how the object is represented as a string.
By defining these dunder methods, we've customized how MyNumber objects are created, added, and displayed.


#### Inheritance

Inheritance allows a class (child class) to inherit attributes and methods from another class (parent class).

```python
class ParentClass:
    def __init__(self, attribute):
        self.attribute = attribute

    def display(self):
        print(f"Attribute: {self.attribute}")

class ChildClass(ParentClass):
    def __init__(self, attribute, child_attribute):
        super().__init__(attribute)  # Call the parent class constructor
        self.child_attribute = child_attribute

    def display_child(self):
        print(f"Child Attribute: {self.child_attribute}")

# Creating an instance of ChildClass
child_object = ChildClass("Parent attribute", "Child attribute")

# Accessing methods from both ParentClass and ChildClass
child_object.display()
child_object.display_child()
```

#### Explanation

- **ParentClass**: A base class with an initializer and a display method.
- **ChildClass**: A derived class that inherits from `ParentClass` and adds an additional attribute and method.
- **super()**: The `super()` function is used to call the constructor of the parent class.

#### Encapsulation and Access Modifiers

Python does not have strict access modifiers like `private`, `protected`, and `public`. However, there are naming conventions to indicate the intended visibility of attributes and methods:

- **Public**: Attributes and methods that can be accessed from outside the class. (No underscores)
- **Protected**: Attributes and methods intended for use within the class and its subclasses. (Single underscore `_`)
- **Private**: Attributes and methods intended for use only within the class. (Double underscore `__`)

```python
class MyClass:
    def __init__(self):
        self.public = "Public"
        self._protected = "Protected"
        self.__private = "Private"

    def display(self):
        print(f"Public: {self.public}")
        print(f"Protected: {self._protected}")
        print(f"Private: {self.__private}")

# Creating an instance of MyClass
obj = MyClass()

# Accessing attributes
print(obj.public)  # Public
print(obj._protected)  # Protected (by convention, should be treated as non-public)
# print(obj.__private)  # AttributeError: 'MyClass' object has no attribute '__private'
obj.display()
```

#### More Examples of Dunder Methods

In Python, the iteration protocol is implemented using two special (“dunder”) methods: **\_\_iter\_\_** and **\_\_next\_\_**. They allow objects to be used in contexts that expect an iterator, such as in `for` loops or with the built-in `next()` function.



## Overview

1. **\_\_iter\_\_()**  
   - Must return an iterator object (usually `self`).  
   - Called when an iteration starts, for example in a `for` loop or by passing an object to the `iter()` function.

2. **\_\_next\_\_()**  
   - Must return the next item in the sequence.  
   - Called repeatedly to get each subsequent item in the iterator.  
   - Must raise `StopIteration` when there are no more items to return.

In summary:

- A class that has an **\_\_iter\_\_()** method returning an iterator object is **iterable**.
- An **iterator** is an object that has both an **\_\_iter\_\_()** method (returning itself) and a **\_\_next\_\_()** method.



## How They Are Used

When you write code like:
```python
for item in some_object:
    ...
```
1. Python calls `some_object.__iter__()` to retrieve an iterator.
2. It then repeatedly calls the iterator’s `__next__()` method to get items one by one.
3. Once the iterator’s `__next__()` method raises `StopIteration`, the `for` loop exits gracefully.

Similarly, the built-in function `iter(some_object)` internally calls `some_object.__iter__()`, and `next(iterator)` internally calls `iterator.__next__()`.

---

### Code Example Building a Simple Iterator

Below is a straightforward example of a custom iterator that yields numbers from 1 up to a specified limit.

```python
class CountUpTo:
    def __init__(self, max):
        self.max = max
        self.current = 0

    def __iter__(self):
        """Return the iterator object (which is self)."""
        # The iteration state is defined by self.current
        return self

    def __next__(self):
        """Return the next item or raise StopIteration."""
        if self.current < self.max:
            self.current += 1
            return self.current
        else:
            raise StopIteration

# Example usage:
counter = CountUpTo(5)
for number in counter:
    print(number)
```

**Explanation:**

- **\_\_iter\_\_**: Returns `self`, because this class instance is its own iterator.  
- **\_\_next\_\_**: Increments `self.current` until it reaches `self.max`. Once `self.current == self.max`, it raises `StopIteration` to signal that the iteration is over.  
- In the `for` loop, each iteration calls `__next__()`. The items printed would be 1, 2, 3, 4, and 5.

### <font color='blue' size=5pt> Vector Spaces - Formal Definition </font>

A vector space $V$ is a set whose elements can be multiplied with scalars and combined among themselves by using a $"\cdot"$ and a $"+"$ operation, respectively.  These operations mentioned above, must satisfy the following properties:

| **Axiom**                                                                                                                                | **Meaning**                                                                                                                                                                                                  |
| ---------------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| [Associativity](https://en.wikipedia.org/wiki/Associativity "Associativity") of vector addition                                          | **u** + (**v** + **w**) = (**u** + **v**) + **w**                                                                                                                                                            |
| [Commutativity](https://en.wikipedia.org/wiki/Commutativity "Commutativity") of vector addition                                          | **u** + **v** = **v** + **u**                                                                                                                                                                                |

| [Inverse elements](https://en.wikipedia.org/wiki/Inverse_element "Inverse element") of vector addition                                   | For every **v**∈ _V_, there exists an element −**v** ∈ _V_, called the _[additive inverse](https://en.wikipedia.org/wiki/Additive_inverse "Additive inverse")_ of **v**, such that **v** + (−**v**) = **0**. |

### <font color='blue' size=5pt> Linear Operators </font>

Formal Definition: A mapping $T$ between two vector spaces is called linear if and only if both statements hold for any scalar $\lambda$ and any two vectors $\vec{u}$ and $\vec{w}$ from the domain of $T.$

> i\. $T\left(\lambda\vec{u}\right) = \lambda T\left(\vec{u}\right)$

> ii\. $T\left(\vec{u}+\vec{v}\right)=T\left(\vec{u}\right)+T\left(\vec{v}\right)$

for any scaler $\lambda$ and any two vectors $\vec{u}$ and $\vec{v}.$


An example of a linear model with three features is how you compute the final grade in a class if you know your grades from homework, midterm and the final, and also know (per class syllabus) their weights of importance.


If $p$ features in the data and we want to create a linear model, the *input-output* mechanism is

$$
\underbrace{Y}_\text{Output}  = \underbrace{w_1 X_1+w_2 X_2+...+w_p X_p}_\text{Linear combination of features}
$$

This could represented as a matrix-vector product:

$$
\underbrace{Y}_\text{Output}  = \underbrace{X\cdot w}_\text{Linear combination of the columns of matrix X}
$$

In this model the features are $X_1, X_2, ...X_p$ and $w_1, w_2,...w_p$ are a set of weights (real numbers).

<font color='deepskyblue'>**Example 4**</font> A Fibonacci sequence.

Iterative approach to generating the Fibonacci sequence, i.e. the sequence $(f_n)$ of integers such that $$f_0=0,\,\, f_1=1\,\,\, \text{and}\,\, f_{n+1}=f_n+f_{n-1}$$ for all $n\geq 1.$

````python
class Fib:
  def __init__(self,a=0,b=1):
    self.a = a
    self.b = b
  def update(self):
    self.a, self.b = self.b, self.a + self.b
    self.current = self.b
````

````python
myfib=Fib(a=0.2,b=1.5)
````

````python
myfib.update()
````

````python
myfib.current
````

````python
myfib.update()
myfib.current
````

####  <font color='deepskyblue' size=5pt>Example: Collatz Conjecture </font>

The Collatz conjecture, also known as the 3n + 1 conjecture, is a mathematical hypothesis that concerns sequences defined as follows:

1. Start with any positive integer $ n $.
2. If $ n $ is even, divide it by 2: $ n \to \frac{n}{2} $.
3. If $ n $ is odd, multiply it by 3 and add 1: $ n \to 3n + 1 $.
4. Repeat the process indefinitely with the resulting value.

The conjecture states that no matter what initial value of $ n $ is chosen, the sequence will eventually reach the number 1. Once it reaches 1, it will enter the repeating cycle of 1, 4, 2, 1.

Despite its simple definition, the Collatz conjecture remains unproven. It has been tested for very large numbers using computers, and no counterexamples have been found. However, a general proof for all positive integers is still elusive, making it one of the most famous unsolved problems in mathematics.

````python
# example with classes
class collatz:
    # attributes initialization
    def __init__(self,x=1):
        self.x = x
    # method definition
    def f(self):
        if self.x%2 == 0:
            return self.x/2
        else:
            return 3*self.x+1
    def __iter__(self):
        return self
    def __next__(self):
        self.x = self.f()
        return self.x
````

````python
myclass = collatz(x=233471)
````

````python
# i is a counter variable
i = 1
while next(myclass)>1:
    i += 1
print('The Collatz conjecture worked in: '+str(i),' iterations.')
````

### <font color='blue' size=6pt>  Scaling Data </font>

<font color='red'> IMPORTANT: If we have to perform calculations with values from different features we must consider scaling the data on the same metric, i.e. we want to add apples to apples.</font>

The are many different techniques for data standardization. We can scale numerical values by using:

  - z-scores: $\frac{x-\mu}{\sigma}$ where $\mu$ is the mean and $\sigma$ is the standard deviation.
  - quantiles: every data point is replaced by its percentile within the feature column.
  - division by max: $\frac{x}{\max{(x)}}.$
  - 0 to 1 scaling: $\frac{x-\min{(x)}}{\max{(x)}-\min{(x)}}.$
  - "traditional" standardization (e.g. old HDI): $\frac{\max{(x)}-x}{\max{(x)}-\min{(x)}}.$
  Example: Life Expectancy Index has minimum age as 25 and maximum age as 85.
Formula of LEI in old method of Human Development Index (HDI) = Life expectancy – 25/ 85 – 25.
  - division by the range.
  - division by the standard deviation.
  - proportion of external indicator (i.e. in terms of a predefined metric).
